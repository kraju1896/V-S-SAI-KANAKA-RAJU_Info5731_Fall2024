Create a Python program that encompasses various string manipulation task:

['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string

# (a) Enter two sentences from the user
sen1 = input("Please enter the first sentence: ")
sen2 = input("Now, enter the second sentence: ")

# (b) Convert sentences into lists of words including punctuation, using regex to split words
import re
words_list = re.findall(r'\w+|[^\w\s]', sen1) + re.findall(r'\w+|[^\w\s]', sen2)
print("Here is the list of words from both sentences:", words_list)

# (c) Find common words between both sentences considered as sets
set_sen1 = set(re.findall(r'\w+', sen1))
set_sen2 = set(re.findall(r'\w+', sen2))
commonwords = list(set_sen1.intersection(set_sen2))
print("These are the common words in both sentences:", commonwords)

# (d) Remove punctuation from sentences provided by user
translated = str.maketrans("", "", string.punctuation)
cleaned_sen1 = sen1.translate(translated)
cleaned_sen2 = sen2.translate(translated)
print("Here is the first sentence after removing punctuation:", cleaned_sen1)
print("Here is the second sentence after removing punctuation:", cleaned_sen2)

# (e) Counting the vowels in each sentence
vowels_considered = "aeiouAEIOU"

def count_vowels(sentence):
    sentence = sentence.lower()  # Convert to lowercase for uniformity
    vowel_counts = {}
    for vowel in vowels_considered:
        count = sentence.count(vowel)
        if count > 0:
            vowel_counts[vowel] = count
    return vowel_counts

vowel_count1 = count_vowels(cleaned_sen1)
vowel_count2 = count_vowels(cleaned_sen2)

print("Vowel counts in the first sentence are as follows:", vowel_count1)
print("Vowel counts in the second sentence are as follows:", vowel_count2)
```

```python
import math

# Function to evaluate a mathematical expression
def evaluate_exp(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except Exception as e:
        return f"Error: Invalid input. {str(e)}"

# Function to run the calculator
def start_calculator():
    print("Welcome to Simple Calculator!")
    print("You can perform these operations: +, -, *, /, ** (exponentiation), % (modulus), sqrt (square root)")
    print("To exit the calculator, just type 'exit'.")

    while True:
        # considering the input from the user
        user_input = input("Enter your mathematical expression: ")

        # Exit condition
        if user_input.lower() == 'exit':
            print("Thanks for using the calculator. Goodbye!")
            break

        # Check if square root operation is requested
        if 'sqrt' in user_input:
            try:
                # Handle square root
                number = float(user_input.split('sqrt')[1].strip(' ()'))
                result = math.sqrt(number)
                print("The result is:", result)
            except ValueError:
                print("Error: Invalid input for square root.")
            except Exception as e:
                print(f"Error: {str(e)}")
        else:
            # For other expressions, evaluate normally
            result = evaluate_exp(user_input)
            print("The result is:", result)

if __name__ == "__main__":
    start_calculator()
```

# Question 5 (20 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import csv

# Load data from CSV file
def load_data(filename):
    data = {}

    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            year = int(row['year'])
            title = row['title_field']
            value = int(row['value'])

            if year not in data:
                data[year] = {'Total Housing Units': 0, 'Vacant Housing Units': 0, 'Occupied Housing Units': 0}

            data[year][title] = value

    return data

# (a) Calculate the difference in the number of Occupied Housing Units from year to year
def calculate_occupied_difference(data):
    differences = []
    years = sorted(data.keys())

    for i in range(1, len(years)):
        year1 = years[i-1]
        year2 = years[i]

        diff = data[year2]['Occupied Housing Units'] - data[year1]['Occupied Housing Units']
        differences.append((year1, year2, diff))

    differences.sort(key=lambda x: x[2])  # Sort by difference value in ascending order
    return differences

# (b) Calculate the percentage of vacant and occupied housing units for each year
def calculate_vacancy_percentage(data):
    percentage_data = []

    for year, values in data.items():
        vacant_percentage = (values['Vacant Housing Units'] / values['Total Housing Units']) * 100
        occupied_percentage = (values['Occupied Housing Units'] / values['Total Housing Units']) * 100
        percentage_data.append((year, round(vacant_percentage, 2), round(occupied_percentage, 2)))

    return percentage_data

# (c) Find the year with the highest number of vacant and occupied housing units
def find_highest_vacancy_occupied(data):
    max_vacant_year = max(data, key=lambda year: data[year]['Vacant Housing Units'])
    max_vacant_value = data[max_vacant_year]['Vacant Housing Units']

    max_occupied_year = max(data, key=lambda year: data[year]['Occupied Housing Units'])
    max_occupied_value = data[max_occupied_year]['Occupied Housing Units']

    return (max_vacant_year, max_vacant_value), (max_occupied_year, max_occupied_value)

# (d) Calculate the harmonic mean of the total housing units
def harmonic_mean(values):
    n = len(values)
    harmonic_sum = sum(1 / value for value in values)
    return n / harmonic_sum

# Main function to process the tasks
def main():
    # Load data from CSV file
    filename = "Assignment1_denton_housing.csv"
    data = load_data(filename)

    # (a) Calculate the difference in occupied housing units year to year
    differences = calculate_occupied_difference(data)
    print("\n(a) Difference in Occupied Housing Units:")
    for year1, year2, diff in differences:
        print(f"From {year1} to {year2}: {diff}")

    # (b) Calculating the percentage of vacant housing units and occupied housing units
    percentage_data = calculate_vacancy_percentage(data)
    print("\n(b) Percentage of Vacant Housing Units and Occupied Housing Units:")
    print(f"{'Year':<10} {'Vacant Housing Units':<20} {'Occupied Housing Units':<20}")
    for year, vacant_percentage, occupied_percentage in percentage_data:
        print(f"{year:<10} {vacant_percentage}% {occupied_percentage}%")

    # (c) Finding the years with the highest vacant units and occupied units
    max_vacant, max_occupied = find_highest_vacancy_occupied(data)
    print("\n(c) Year and value for highest vacant units and occupied units:")
    print(f"Number of Vacant Housing Units - Year: {max_vacant[0]}, Value: {max_vacant[1]}")
    print(f"Number of Occupied Housing Units - Year: {max_occupied[0]}, Value: {max_occupied[1]}")

    # (d) Calculating the harmonic mean of total housing units
    total_housing_units = [data[year]['Total Housing Units'] for year in data]
    harmonic = harmonic_mean(total_housing_units)
    print(f"\n(d) Harmonic mean of total housing units: {round(harmonic, 2)}")

if __name__ == "__main__":
    main()
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os

# Constants
INVENTORY_FILE = "inventory.csv"
RESTOCK_THRESHOLD = 5  # Auto-restock when quantity drops below this level
RESTOCK_QUANTITY = 20  # Refill to this amount when restocking

# Initialize inventory with sample products (run once)
def initialize_inventory():
    if not os.path.exists(INVENTORY_FILE):
        with open(INVENTORY_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])
            for i in range(1, 51):
                writer.writerow([i, f"Product{i}", "Category" + str(i % 5), round(10 + i * 0.5, 2), 10])

def load_inventory():
    inventory = []
    try:
        with open(INVENTORY_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                try:
                    row['ID'] = int(row['ID'])
                    row['Price'] = float(row['Price'])
                    row['Quantity'] = int(row['Quantity'])
                    inventory.append(row)
                except ValueError:
                    print(f"Skipping corrupt row: {row}")
    except FileNotFoundError:
        print("Inventory file not found. Initializing...")
        initialize_inventory()
    return inventory

# Save inventory to CSV
def save_inventory(inventory):
    with open(INVENTORY_FILE, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])
        for item in inventory:
            writer.writerow([item['ID'], item['Name'], item['Category'], item['Price'], item['Quantity']])

# Add a new product
def add_product(id, name, category, price, quantity):
    inventory = load_inventory()
    if any(item['ID'] == id for item in inventory):
        print("Error: Product ID already exists.")
        return
    inventory.append({"ID": id, "Name": name, "Category": category, "Price": price, "Quantity": quantity})
    save_inventory(inventory)
    print("Product added successfully.")

# Delete a product
def delete_product(id):
    inventory = load_inventory()
    if not any(item['ID'] == id for item in inventory):
        print("Error: Product not found.")
        return
    inventory = [item for item in inventory if item['ID'] != id]
    save_inventory(inventory)
    print("Product deleted.")

# Update product details
def update_product(id, new_quantity=None, new_price=None):
    inventory = load_inventory()
    for item in inventory:
        if item['ID'] == id:
            if new_quantity is not None:
                if new_quantity < 0:
                    print("Error: Quantity cannot be negative.")
                    return
                item['Quantity'] = new_quantity
            if new_price is not None:
                item['Price'] = new_price
            save_inventory(inventory)
            print("Product updated successfully.")
            return
    print("Error: Product not found.")

# Search products
def search_product(name=None, category=None):
    inventory = load_inventory()
    results = [item for item in inventory if (name and name.lower() in item['Name'].lower()) or (category and category.lower() == item['Category'].lower())]
    return results

# Auto-restocking system
def restock_inventory():
    inventory = load_inventory()
    for item in inventory:
        if item['Quantity'] < RESTOCK_THRESHOLD:
            item['Quantity'] = RESTOCK_QUANTITY
    save_inventory(inventory)
    print("Restock completed.")

# Generate category breakdown report
def generate_report():
    inventory = load_inventory()
    category_summary = {}
    for item in inventory:
        category_summary[item['Category']] = category_summary.get(item['Category'], 0) + item['Price'] * item['Quantity']
    print("Category Breakdown:")
    for category, value in category_summary.items():
        print(f"{category}: ${value:.2f}")

# Edge Cases Testing
def run_edge_cases():
    print("Running edge cases...")
    add_product(55, "NewProduct", "Category1", 25.0, 10)  # Adding a new product
    add_product(1, "DuplicateProduct", "Category1", 20.0, 5)  # Duplicate ID
    delete_product(55)  # Deleting a product
    delete_product(100)  # Non-existent product ID
    update_product(1, new_quantity=15)  # Updating a product
    update_product(100, new_quantity=50)  # Updating a product that doesn’t exist
    update_product(5, new_quantity=-10)  # Negative quantity
    print(search_product("NonExistentProduct"))  # Search for non-existent product
    restock_inventory()  # Test auto-restocking
    generate_report()  # Check category breakdown

# Uncomment the following line to initialize inventory if running for the first time
initialize_inventory()

# Running tests
run_edge_cases()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
import csv
import os

# Constants
FLIGHT_FILE = "flights.csv"
SEAT_CAPACITY = 150  # Default seat capacity per flight

# Initialize flights with sample data
def initialize_flights():
    if not os.path.exists(FLIGHT_FILE):
        with open(FLIGHT_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable"])
            for i in range(1, 51):
                writer.writerow([i, f"City{i}", f"City{i+1}", f"2025-07-{i%30+1} 10:00", SEAT_CAPACITY])

# Load flight data
def load_flights():
    flights = []
    try:
        with open(FLIGHT_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                try:
                    row['FlightID'] = int(row['FlightID'])
                    row['SeatsAvailable'] = int(row['SeatsAvailable'])
                    flights.append(row)
                except ValueError:
                    print(f"Skipping corrupt row: {row}")
    except FileNotFoundError:
        print("Flight file not found. Initializing...")
        initialize_flights()
    return flights

# Save flight data
def save_flights(flights):
    with open(FLIGHT_FILE, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable"])
        for flight in flights:
            writer.writerow([flight['FlightID'], flight['Origin'], flight['Destination'], flight['DepartureTime'], flight['SeatsAvailable']])

# Add a new flight
def add_flight(flight_id, origin, destination, departure_time, seats_available):
    flights = load_flights()
    if any(flight['FlightID'] == flight_id for flight in flights):
        print("Error: Flight ID already exists.")
        return
    flights.append({"FlightID": flight_id, "Origin": origin, "Destination": destination, "DepartureTime": departure_time, "SeatsAvailable": seats_available})
    save_flights(flights)
    print("Flight added successfully.")

# Delete a flight
def delete_flight(flight_id):
    flights = load_flights()
    flights = [flight for flight in flights if flight['FlightID'] != flight_id]
    save_flights(flights)
    print(f"Flight {flight_id} deleted successfully.")

# Find a flight
def find_flights(origin=None, destination=None):
    flights = load_flights()
    results = [flight for flight in flights if (origin and origin.lower() in flight['Origin'].lower()) or (destination and destination.lower() in flight['Destination'].lower())]
    return results

# Reserve a seat
def reserve_seat(flight_id):
    flights = load_flights()
    for flight in flights:
        if flight['FlightID'] == flight_id:
            if flight['SeatsAvailable'] > 0:
                flight['SeatsAvailable'] -= 1
                save_flights(flights)
                print("Seat reserved successfully.")
                return
            else:
                print("Error: No available seats.")
                return
    print("Error: Flight not found.")

# Release a seat
def release_seat(flight_id):
    flights = load_flights()
    for flight in flights:
        if flight['FlightID'] == flight_id:
            flight['SeatsAvailable'] += 1
            save_flights(flights)
            print("Seat released successfully.")
            return
    print("Error: Flight not found.")

# Generate performance report
def generate_report():
    flights = load_flights()
    total_seats = sum(flight['SeatsAvailable'] for flight in flights)
    print(f"Total Available Seats: {total_seats}")
    print("Flight Breakdown:")
    for flight in flights:
        print(f"Flight {flight['FlightID']}: {flight['SeatsAvailable']} seats available")

# Edge Cases Testing
def run_edge_cases():
    print("Running edge cases...")
    add_flight(53, "NewYork", "LosAngeles", "2025-07-15 08:00", 200)  # Adding a new flight
    add_flight(1, "DuplicateCity", "DuplicateCity", "2025-07-10 10:00", 150)  # Duplicate ID
    reserve_seat(1)  # Reserving a seat
    reserve_seat(100)  # Non-existent flight
    release_seat(1)  # Releasing a seat
    release_seat(100)  # Releasing a seat on a non-existent flight
    reserve_seat(1)  # Overbooking test (should not go below 0)
    print(find_flights(origin="City5"))  # Search by origin
    generate_report()  # Check flight performance report
    delete_flight(53)  # Deleting a flight

# Initialize and run tests
initialize_flights()
run_edge_cases()
```
